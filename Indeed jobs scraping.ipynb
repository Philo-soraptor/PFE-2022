```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
```

```python
url="https://ma.indeed.com/emplois?q=data&l=maroc&lang=fr"

options = webdriver.ChromeOptions()
#options.add_argument('--headless')
chrome_path = 'C:/Users/socrates/desktop/chromedriver.exe'
driver = webdriver.Chrome(chrome_path,options=options)
driver.maximize_window()
driver.get(url)
```

```python
id_link = 1
id_td = 1

job_links = []
job_titles = []
job_descriptions = []

for i in range(5):
    link = url + "&start=" + str(i*10)
    driver.get(link)
    time.sleep(3)
    try:
        ClosePopUp = WebDriverWait(driver, 3).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='popover-x']/button"))).click()
    except:
        pass
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    jobs = driver.find_elements_by_xpath("//td[@class='resultContent']/div[1]/h2[1]/a[1]")
    for job in jobs:
        job_link = job.get_attribute('href')
        job_links.append(job_link)
        print("link N°", id_link, ": saved")
        id_link += 1

for job_link in job_links:  
    try:
        driver.get(job_link)
        time.sleep(5)
        
        title = driver.find_element_by_class_name('icl-u-xs-mb--xs').get_attribute('innerText')
        job_titles.append(title)
        print("title N°", id_td, ": saved")
        job_description = driver.find_element_by_xpath('//div[@id="jobDescriptionText"]').get_attribute('innerText')
        job_descriptions.append(job_description)
        print("jobDescriptionText N°", id_td, ": saved")
        id_td += 1
    except:
        pass
```

```python
column_names = ['Title','Description','Link']

job_data = pd.DataFrame(columns = column_names)

job_data['Title'] = job_titles
job_data['Link'] = job_links
job_data['Description'] = job_descriptions

print(len(job_data))

job_data['Description'] = job_data['Description'].replace('\n',' ')

job_data.to_excel("indeed.xlsx",index = False)   
```
