```python
# install selenium first
# selenium imports
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
import urllib.request
import time

options = webdriver.ChromeOptions()
options.add_experimental_option('prefs', {
"download.default_directory": "C:\\Users\\Socrates\\Desktop\\cv", #Change default directory for downloads
"download.prompt_for_download": False, #To auto download the file
"download.directory_upgrade": True,
"plugins.always_open_pdf_externally": True #It will not show PDF directly in chrome
})
driver = webdriver.Chrome('C:/Users/Socrates/Desktop/chromedriver.exe',options=options) #Change the path of chromedriver.exe

driver.get("https://www.linkedin.com/")
driver.maximize_window()
time.sleep(3)

username = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='session_key']")))
password = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='session_password']")))

username.send_keys("e-scraper@protonmail.com")
password.send_keys("cvscraping2022")

log_in = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']"))).click()
time.sleep(3)
```

```python
keywords = ["INSEA CV","ENSIAS CV","INPT CV"]
for keyword in keywords :
    driver.get("https://www.linkedin.com/")
    time.sleep(3)
    search = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[class='search-global-typeahead__input always-show-placeholder']")))
    search.clear()
    search.send_keys(keyword)
    search.send_keys(Keys.ENTER)
    time.sleep(3)
    url = driver.current_url
    driver.get("https://www.linkedin.com/search/results/content/?keywords=" + url[54:])
    time.sleep(3)
    Flag = True
    
    Number_Of_Docs = -1
    while Flag :
        action = ActionChains(driver)
        for i in range(80):
            action.send_keys(Keys.PAGE_DOWN)
            action.perform()
            time.sleep(0.75)
        iframe = driver.find_elements_by_xpath("//iframe[@class='document-s-container__document-element document-s-container__document-element--loaded']") 
        nDocs = len(iframe)
        if Number_Of_Docs == nDocs :
            Flag = False
            print('There are no more document with keyword :',keyword)
        else :
            print('There are more document with keyword :',keyword)
            Number_Of_Docs = nDocs

    print('There are',Number_Of_Docs, 'documents with keyword :',keyword)

    docs_list = []

    for i in range(Number_Of_Docs):
        driver.switch_to.frame(iframe[i])
        document = driver.find_element_by_class_name('ssplayer-virus-scan-container__download-button')
        link = document.get_attribute('href')
        print('link',i+1,link,"\n")
        driver.get(link)
        time.sleep(3)
        docs_list.append((link,keyword))
        driver.switch_to.default_content()

    docs_list
    
```
